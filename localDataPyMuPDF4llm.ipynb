```python
import os
INPUT_FOLDER = "papers"
OUTPUT_FOLDER = "papers_markdown"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)
```

```python
import pymupdf4llm
from tqdm import tqdm
import pathlib

pdf_files = [f for f in os.listdir(INPUT_FOLDER) if f.lower().endswith('.pdf')]
 
for filename in tqdm(pdf_files, desc="Converting to MD"):
    pdf_path = os.path.join(INPUT_FOLDER, filename)
    md_out_path = os.path.join(OUTPUT_FOLDER, filename.replace(".pdf", ".md"))

    if os.path.exists(md_out_path):
        continue

    try:
        # pymupdf4llm converts the PDF to a Markdown string
        md_text = pymupdf4llm.to_markdown(pdf_path)
        
        with open(md_out_path, "w", encoding="utf-8") as f:
            f.write(md_text)
    except Exception as e:
        print(f"Error on {filename}: {e}")
```

```python
import re
import glob
import jsonlines

TEMP_JSONL = "cleaned_temp1.jsonl"

def clean_text(text):
    # Split at common bibliography headers (case insensitive)
    pattern = re.compile(r'\n(?:References|Bibliography|Works Cited)\n', re.IGNORECASE)
    parts = pattern.split(text)
    text = parts[0]
    
    # Clean up whitespace
    text = re.sub(r'\n{3,}', '\n\n', text)
    return text.strip()

md_files = glob.glob(f"{OUTPUT_FOLDER}/*.md")

with jsonlines.open(TEMP_JSONL, mode='w') as writer:
    for file_path in tqdm(md_files, desc="Cleaning"):
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()
            
        cleaned = clean_text(content)
        
        # Keep papers with at least 1500 characters
        if len(cleaned) > 1500:
            writer.write({
                "text": cleaned,
                "metadata": {"source": os.path.basename(file_path)}
            })

print(f"Cleaning complete. Temporary data saved to {TEMP_JSONL}")
```

```python
import re
import glob
import jsonlines
import ftfy
import langid
from tqdm import tqdm

TEMP_JSONL = "cleaned_temp1.jsonl"

def robust_clean(text):
    # 1. Fix Unicode/Ligatures (e.g., "ﬁ" -> "fi", "ﬂ" -> "fl")
    text = ftfy.fix_text(text)

    # 2. Strip Bibliography
    # Split at common bibliography headers (case insensitive)
    bib_pattern = re.compile(r'\n(?:References|Bibliography|Works Cited|REFERENCE)\n', re.IGNORECASE)
    text = bib_pattern.split(text)[0]

    # 3. Remove Citations
    # Removes [1], [1, 2], or (Smith et al., 2020) to make prose smoother
    text = re.sub(r'\[\d+(?:,\s*\d+)*\]', '', text) # [1, 2]
    text = re.sub(r'\([A-Z][a-z]+(?:\set\sal\.)?,\s\d{4}\)', '', text) # (Smith et al., 2020)

    # 4. Remove boilerplate headers/footers
    # Strips things like "Page 1 of 10" or "arXiv:1234.5678"
    text = re.sub(r'(?i)page\s\d+\s(of|/)\s\d+', '', text)
    text = re.sub(r'(?i)arxiv:\d+\.\d+(v\d+)?', '', text)

    # 5. Normalize whitespace
    text = re.sub(r'\s+', ' ', text) # Replace all tabs/newlines with single space
    text = re.sub(r'\n{2,}', '\n\n', text) # Keep paragraph breaks
    
    return text.strip()

def is_high_quality(text):
    """Heuristic checks to see if the text is worth keeping."""
    if len(text) < 1000: return False # Too short
    
    # Check Alphanumeric density
    # If a doc is 40% math symbols/tables, it's bad for general prose training
    alnum_count = sum(1 for char in text if char.isalnum())
    if alnum_count / len(text) < 0.7: return False 
    
    # Language check (Ensures only English papers are included)
    lang, score = langid.classify(text)
    if lang != 'en': return False
    
    return True

md_files = glob.glob(f"{OUTPUT_FOLDER}/*.md")

with jsonlines.open(TEMP_JSONL, mode='w') as writer:
    for file_path in tqdm(md_files, desc="Robust Cleaning"):
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            
            cleaned = robust_clean(content)
            
            if is_high_quality(cleaned):
                writer.write({
                    "text": cleaned,
                    "metadata": {
                        "source": os.path.basename(file_path),
                        "char_count": len(cleaned)
                    }
                })
        except Exception as e:
            continue 

print(f"\nRobust cleaning complete. Results in {TEMP_JSONL}")
```

```python
from datasketch import MinHash, MinHashLSH

FINAL_JSONL = "research_dataset_final1.jsonl"
lsh = MinHashLSH(threshold=0.85, num_perm=128)
unique_count = 0

with jsonlines.open(TEMP_JSONL) as reader, jsonlines.open(FINAL_JSONL, mode='w') as writer:
    for i, obj in enumerate(tqdm(reader, desc="Deduplicating")):
        m = MinHash(num_perm=128)
        # Create shingles from words
        for word in set(obj['text'].lower().split()):
            m.update(word.encode('utf8'))
        
        if not lsh.query(m):
            lsh.insert(f"doc_{i}", m)
            writer.write(obj)
            unique_count += 1

print(f"Success! Final unique papers: {unique_count}")
```
